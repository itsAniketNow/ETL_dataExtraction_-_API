#Download Last Week Data

```python
# Calculate the date one week ago
last_week_date = (datetime.now() - timedelta(weeks=1)).strftime("%Y-%m-%dT%H:%M:%S")

# Spotify API endpoint to get recently played tracks
recently_played_url = "https://api.spotify.com/v1/me/player/recently-played"
headers = {"Authorization": f"Bearer {access_token}"}
params = {"limit": 50, "after": last_week_date}

# Make request to get recently played tracks
response = requests.get(recently_played_url, headers=headers, params=params)
recently_played_data = response.json()

# Process the data as needed (e.g., store in a database, perform transformations)

# Example: Print the track names
#for item in recently_played_data["items"]:
    #track_name = item["track"]["name"]
    #print(f"Track: {track_name}")

# Print the extracted data
print("Extracted Data:")
print(recently_played_data)

# Example: Print specific information (e.g., track names)
for item in recently_played_data["items"]:
    track_name = item["track"]["name"]
    artist_name = item["track"]["artists"][0]["name"]
    played_at = item["played_at"]
    print(f"Track: {track_name}, Artist: {artist_name}, Played At: {played_at}")
```

#Validation of Data

```python
# Validate data format
def validate_data_format(item):
    required_fields = ["track", "played_at"]
    for field in required_fields:
        if field not in item:
            return False, f"Missing required field: {field}"

    # Check data types
    if not isinstance(item["track"], dict):
        return False, "Invalid data type for 'track'"

    # Check track name
    if not item["track"].get("name"):
        return False, "Empty track name"

    # Check timestamp format
    try:
        datetime.strptime(item["played_at"], "%Y-%m-%dT%H:%M:%S.%fZ")
    except ValueError:
        return False, "Invalid timestamp format for 'played_at'"

    return True, "Data format is valid"

# Remove duplicates
def remove_duplicates(data):
    unique_data = []
    seen_items = set()

    for item in data:
        # Convert the dictionary to a JSON-formatted string
        item_json = json.dumps(item, sort_keys=True)
        if item_json not in seen_items:
            seen_items.add(item_json)
            unique_data.append(item)

    return unique_data

# Validate and process the data
for item in recently_played_data["items"]:
    is_valid, message = validate_data_format(item)
    if not is_valid:
        print(f"Invalid data format: {message}")

# Remove duplicates
unique_data = remove_duplicates(recently_played_data["items"])

# Print the unique track names
for item in unique_data:
    track_name = item["track"]["name"]
    print(f"Track: {track_name}")
```
